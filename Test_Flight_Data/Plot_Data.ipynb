```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("flightData_0_deg.csv")
# Convert Time to seconds
data["Time"] = data["Time"]/1000

plt.plot(data["Time"], data["Altitude"])
plt.ylabel('Altitude (m)')
plt.xlabel('Time (s)')
plt.show()

plt.plot(data["Time"], data["Velocity"])
plt.ylabel('Velocity (m/s)')
plt.xlabel('Time (s)')
plt.show()

plt.plot(data["Time"], data["AccX"])
plt.ylabel('Vertical Acceleration (m/s)')
plt.xlabel('Time (s)')
plt.show()

plt.plot(data["Time"], data["Apogee Predict"])
plt.ylabel('Apogee Predict (m)')
plt.xlabel('Time (s)')
plt.show()
```
